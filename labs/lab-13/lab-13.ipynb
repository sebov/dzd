```python
import deeplake

# Create a local dataset
ds = deeplake.create("data/dataset1")

ds = deeplake.open("data/dataset1")
```

```python
ds.add_column("ids", "int32")
ds.add_column("labels", "text")
```

```python
ds.append({
    "ids": [1, 2, 3],
    "labels": ["cat", "dog", "bird"],
})
ds.commit()
```
